```python
index = 0
projnextweek_pointssofar2020_lastweekspoints = []
for i in soup.findAll('div', {"class": "jsx-2810852873 table--cell total tar"}):
    string = str(i).replace('>', '<').split('<')[4]
    if index % 3 == 0 and index != 0:
        print(projnextweek_pointssofar2020_lastweekspoints)
        projnextweek_pointssofar2020_lastweekspoints = []
    projnextweek_pointssofar2020_lastweekspoints.append(string)
    index+= 1
    
```

```python
#Putting all together

players = []
for i in soup.findAll("a", {"class": "AnchorLink link clr-link pointer"}):
    players.append(str(i).replace('>', '<').split('<')[2])
    
pos = []
for i in soup.findAll('div', {"class": "jsx-2810852873 table--cell"}):
    string = str(i)
    if 'title' in string and 'points' not in string:
        pos.append(string.replace('>', '<').split('<')[2])
        
stats = []
index = 0
projnextweek_pointssofar2020_lastweekspoints = []
for i in soup.findAll('div', {"class": "jsx-2810852873 table--cell total tar"}):
    string = str(i).replace('>', '<').split('<')[4]
    if index % 3 == 0 and index != 0:
        stats.append(projnextweek_pointssofar2020_lastweekspoints)
        projnextweek_pointssofar2020_lastweekspoints = []
    projnextweek_pointssofar2020_lastweekspoints.append(string)
    index+= 1
    
projnextweek,pointssofar2020,lastweekspoints = [],[],[]
for i in stats:
    projnextweek.append(i[0])
    pointssofar2020.append(i[1])
    lastweekspoints.append(i[2])
    
pd.DataFrame({"Players": players[0:20], "Position": pos[0:20],
              "Week Projection": projnextweek,
              "2020 Points": pointssofar2020,
              "Prev Week Points": lastweekspoints})
```
